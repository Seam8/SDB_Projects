```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
Data = pd.read_csv(filepath_or_buffer ='selim_bourail_station.txt', sep=';', header=0,decimal=',', parse_dates=['DATE'])
'''Data['DATE']= pd.to_datetime(Data['DATE'], format='%Y%m%d')'''
Data.index= Data['DATE']; del Data['DATE']
print(len(Data))
Data.isnull().sum()
Data
```

```python
Data['2018-03-01':'2018-04-01'].plot(kind='line',y='RR')
```

```python
CumData = Data['2018-03-01':'2018-04-01'] + Data['2018-03-02':'2018-04-02'] + Data['2018-03-03':'2018-04-03']
CumData['RR']
```

```python
fig, ax = plt.subplots(figsize=(8,7))
#set major ticks format
TimeStamp =  ['2018-03-01','2018-04-01']
ax.xaxis.set_ticks(Data[TimeStamp[0]:TimeStamp[1]].index)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m'))
ax.bar(Data[TimeStamp[0]:TimeStamp[1]].index, Data[TimeStamp[0]:TimeStamp[1]]['RR'].rolling(4).sum())
ax.xaxis_date()
plt.setp(ax.get_xticklabels(), rotation=90)
plt.show
```
