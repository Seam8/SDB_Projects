```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
weatherFileName =  'D:/SDB/Meteo/data_stations_kethsia/data_station_Noumea.csv'

Data = pd.read_csv(filepath_or_buffer =weatherFileName,  encoding = "latin1", sep=';', header=0,decimal=',')
Data['DATE']= pd.to_datetime(Data['DATE'],format='%Y%m%d%H')
Data.index= Data['DATE']; del Data['DATE']
Data = Data.tz_localize(tz='Pacific/Noumea').tz_convert('UTC')
print(len(Data))
print(Data.isnull().sum())
Data
```
